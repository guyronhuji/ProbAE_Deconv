```python
# Marker UMAP from stored fits only (no model retraining)
import sys
import numpy as np
import pandas as pd
import anndata as ad
import scipy.sparse as sp
import os

os.environ.setdefault("MPLCONFIGDIR", str((REPO_ROOT / "comparisons" / ".mplconfig").resolve()))
os.environ.setdefault("NUMBA_CACHE_DIR", str((REPO_ROOT / "comparisons" / ".numba_cache").resolve()))
(REPO_ROOT / "comparisons" / ".mplconfig").mkdir(parents=True, exist_ok=True)
(REPO_ROOT / "comparisons" / ".numba_cache").mkdir(parents=True, exist_ok=True)

import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
from sklearn.decomposition import PCA
from IPython.display import Image, display

if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))
if str(REPO_ROOT / 'src') not in sys.path:
    sys.path.insert(0, str(REPO_ROOT / 'src'))

from cytof_archetypes.evaluation.embeddings import umap_fit_transform_large
from cytof_archetypes.evaluation.deconvolution_metrics import per_cell_weight_entropy

METHODS = [
    "nmf",
    "classical_archetypes",
    "deterministic_archetypal_ae",
    "probabilistic_archetypal_ae",
]
DISPLAY_NAMES = {
    "nmf": "NMF",
    "classical_archetypes": "Classical Archetypal Analysis",
    "deterministic_archetypal_ae": "Deterministic Archetypal AE",
    "probabilistic_archetypal_ae": "Probabilistic Archetypal AE",
}

method_summary = pd.read_csv(OUT_DIR / "tables" / "levine32_method_summary.csv")
k_sweep = pd.read_csv(OUT_DIR / "tables" / "levine32_k_sweep_summary.csv")

recommended_k = {}
if "recommended_k" in k_sweep.columns:
    recommended_k = k_sweep.groupby("method", as_index=False)["recommended_k"].first().set_index("method")["recommended_k"].to_dict()

rep = {}
for method in METHODS:
    sub = method_summary[method_summary["method"] == method].copy()
    if sub.empty:
        continue
    if method in recommended_k:
        sub = sub[sub["k"] == int(recommended_k[method])]
    if sub.empty:
        continue
    row = sub.sort_values(["val_mse", "seed"], ascending=[True, True]).iloc[0]
    rep[method] = {"k": int(row["k"]), "seed": int(row["seed"])}

if not rep:
    raise RuntimeError("No stored runs were found in method summary.")

# Use one representative run to define the shared test-cell order.
ref_method = next(iter(rep))
ref_spec = rep[ref_method]
ref_weights_path = OUT_DIR / "runs" / ref_method / f"dim_{ref_spec['k']:02d}" / f"seed_{ref_spec['seed']}" / "test" / "weights.csv"
ref_weights = pd.read_csv(ref_weights_path)
test_cell_ids = ref_weights["cell_id"].astype(str).to_numpy()

adata = ad.read_h5ad(OUT_DIR / "datasets" / "levine32_labeled_processed.h5ad")
if "cell_id" in adata.obs.columns:
    adata_cell_ids = adata.obs["cell_id"].astype(str)
else:
    adata_cell_ids = pd.Series(adata.obs_names.astype(str), index=adata.obs_names)

lookup = pd.Series(np.arange(adata.n_obs, dtype=np.int64), index=adata_cell_ids.to_numpy())
idx = lookup.reindex(test_cell_ids)
if idx.isna().any():
    missing = int(idx.isna().sum())
    raise ValueError(f"Could not align {missing} test cells from stored weights to processed dataset.")
idx = idx.to_numpy(dtype=np.int64)

x_all = adata.X
if sp.issparse(x_all):
    x_test = x_all[idx].toarray().astype(np.float32)
else:
    x_test = np.asarray(x_all[idx], dtype=np.float32)

try:
    marker_umap = umap_fit_transform_large(
        x_test,
        n_components=2,
        n_neighbors=25,
        min_dist=0.2,
        random_state=321,
    )
except Exception:
    pca = PCA(n_components=2, random_state=321)
    marker_umap = pca.fit_transform(x_test).astype(np.float32)

method_rows = [m for m in METHODS if m in rep]
fig, axes = plt.subplots(len(method_rows), 2, figsize=(14, 4.2 * len(method_rows)))
if len(method_rows) == 1:
    axes = np.expand_dims(axes, axis=0)

overlay_rows = []
for row_idx, method in enumerate(method_rows):
    spec = rep[method]
    w_path = OUT_DIR / "runs" / method / f"dim_{spec['k']:02d}" / f"seed_{spec['seed']}" / "test" / "weights.csv"
    wdf = pd.read_csv(w_path)

    # Keep a shared row order across methods using the same test-cell ids.
    if not np.array_equal(wdf["cell_id"].astype(str).to_numpy(), test_cell_ids):
        wdf = wdf.set_index("cell_id").reindex(test_cell_ids).reset_index()

    w_cols = [c for c in wdf.columns if c.startswith("w_")]
    w = wdf[w_cols].to_numpy(dtype=np.float32)
    w = w / np.clip(w.sum(axis=1, keepdims=True), 1e-8, None)

    dominant = np.argmax(w, axis=1).astype(int)
    effective_k = np.exp(per_cell_weight_entropy(w)).astype(np.float32)

    ax_dom = axes[row_idx, 0]
    ax_eff = axes[row_idx, 1]

    s1 = ax_dom.scatter(marker_umap[:, 0], marker_umap[:, 1], c=dominant, s=5, alpha=0.75, cmap="tab10")
    ax_dom.set_title(f"{DISPLAY_NAMES.get(method, method)} | marker UMAP: dominant component/archetype (K={spec['k']}, seed={spec['seed']})")
    ax_dom.set_xlabel("UMAP-1")
    ax_dom.set_ylabel("UMAP-2")
    fig.colorbar(s1, ax=ax_dom, fraction=0.046, pad=0.04)

    s2 = ax_eff.scatter(
        marker_umap[:, 0],
        marker_umap[:, 1],
        c=effective_k,
        s=5,
        alpha=0.75,
        cmap="viridis",
        norm=Normalize(vmin=1.0, vmax=float(spec["k"])),
    )
    ax_eff.set_title(f"{DISPLAY_NAMES.get(method, method)} | marker UMAP: effective K")
    ax_eff.set_xlabel("UMAP-1")
    ax_eff.set_ylabel("UMAP-2")
    fig.colorbar(s2, ax=ax_eff, fraction=0.046, pad=0.04)

    labels = wdf["label"].astype(str).to_numpy() if "label" in wdf.columns else np.asarray([""] * len(wdf), dtype=object)
    for i in range(len(wdf)):
        overlay_rows.append(
            {
                "method": method,
                "seed": int(spec["seed"]),
                "k": int(spec["k"]),
                "cell_id": str(test_cell_ids[i]),
                "label": str(labels[i]),
                "umap_1": float(marker_umap[i, 0]),
                "umap_2": float(marker_umap[i, 1]),
                "dominant_archetype": int(dominant[i]),
                "effective_k": float(effective_k[i]),
            }
        )

fig.tight_layout()
fig_path = OUT_DIR / "figures" / "figure7_marker_umap_effective_k_and_dominant_archetypes.png"
fig.savefig(fig_path, dpi=220)
plt.show()
plt.close(fig)

overlay_df = pd.DataFrame(overlay_rows)
overlay_path = OUT_DIR / "tables" / "levine32_marker_umap_dominant_effective_k.csv"
overlay_df.to_csv(overlay_path, index=False)

print("Saved figure:", fig_path)
print("Saved table:", overlay_path)
print("Representative runs used:")
print(pd.DataFrame([
    {"method": m, "k": rep[m]["k"], "seed": rep[m]["seed"]}
    for m in method_rows
]).to_string(index=False))

display(Image(filename=str(fig_path), width=1200))
```
